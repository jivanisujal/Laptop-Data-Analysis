```python
laptop['I3']=laptop['CPU'].apply(lambda x:1 if 'Intel Core i3' in x else 0)
laptop['I5']=laptop['CPU'].apply(lambda x:1 if 'Intel Core i5' in x else 0)
laptop['I7']=laptop['CPU'].apply(lambda x:1 if 'Intel Core i7' in x else 0)
laptop
```

```python
def processor(x):
    if x=="Intel Core i5" or  x=="Intel Core i7" or x=="Intel Core i3":
        return x
    elif x.split()[0]=='Intel':
        return 'Other Intel processor'
    else:
        return 'AMD Processor'
```

```python
laptop['cpu brand']=laptop['CPU'].apply(processor)
```

```python
x=laptop[['Company','TypeName','Inches','ScreenResolution','Cpu','Ram','Gpu','OpSys','Weight','Memory_size','Memory_type']]
x
```

```python
y=laptop[['Price']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline 
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['Company','TypeName','ScreenResolution','Cpu','Ram','Gpu','OpSys','Weight','Memory_size','Memory_type']])
```

```python
column_transfer=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Company','TypeName','ScreenResolution','Cpu','Ram','Gpu','OpSys','Weight','Memory_size','Memory_type']),remainder='passthrough')
column_transfer
```

```python
lr=LinearRegression()
```

```python
pipe_line=make_pipeline(column_transfer,lr)
pipe_line
```

```python
pipe_line.fit(x_train,y_train)
```
